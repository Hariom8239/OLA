```python
mean_ride_distance = data["Ride_Distance"].median()
data["Ride_Distance"].fillna(value=mean_ride_distance, inplace=True)
```

```python
mean_booking_value = data["Booking_Value"].median()
data["Booking_Value"].fillna(value=mean_booking_value,inplace=True)
```

```python
mean_driver_ratings = data["Driver_Ratings"].median()
data["Driver_Ratings"].fillna(mean_driver_ratings,inplace = True)
```

```python
mean_customer_rating = data["Customer_Rating"].median()
data["Customer_Rating"].fillna(mean_customer_rating, inplace=True)
```

```python
mode_payment_method = data["Payment_Method"].mode()[0]
data["Payment_Method"].fillna(mode_payment_method, inplace = True)
```

```python
data["Incomplete_Rides"].value_counts()
```

```python
mask = data["Incomplete_Rides"].isna() # Only take where value is nan

# To fill nan values using another feature i.e. Booking_Status, if its value = Success then fill No , otherwise Yes
data.loc[mask & (data["Booking_Status"]== "Success"),"Incomplete_Rides"] ="No"

# fill the remaining Nan values with No
data.loc[mask, "Incomplete_Rides"] = "Yes"
```

```python
#Lets fill the missing values in Cancelede_by_Driver feature using another feature i.e. Booking_Status 
mask = data["Canceled_Rides_by_Driver"].isna()

# Now we will fill the Nan values as Not canceled if Booking_Status is Success.
data.loc[mask & (data["Booking_Status"]=="Success"), "Canceled_Rides_by_Driver"] = "Not canceled"

# In case when driver is not available , set the nan value to customer not found.
data.loc[mask & (data["Booking_Status"]=="Driver Not Found"), "Canceled_Rides_by_Driver"] = "Customer not found"

# Fill remaining nan values with Canceled by customer
data.loc[mask & (data["Booking_Status"]=="Canceled by Customer"), "Canceled_Rides_by_Driver"] = "Canceled by Customer"
```

Lets replace the NaN values in our column "Incomplete_Rides_Reason" with the help of other columns such as "Canceled_Rides_by_Customer" and "Canceled_Rides_by_Driver". 

```python
# Boolean masking for replacing only the NaN values.
mask = data["Incomplete_Rides_Reason"].isna()

# We will replace the NaN values using the values of our two other features. First by using feature 'Canceled_Rides_by_Customer':
data.loc[mask & (data['Canceled_Rides_by_Customer'] == 'Driver is not moving towards pickup location'), "Incomplete_Rides_Reason"] = 'Driver is not moving towards pickup location '
data.loc[mask & (data['Canceled_Rides_by_Customer'] == 'Driver asked to cancel'), 'Incomplete_Rides_Reason'] = 'Driver asked Customer to cancel'
data.loc[mask & (data['Canceled_Rides_by_Customer'] == 'Change of plans'), 'Incomplete_Rides_Reason'] = 'Change of plans of Customer'
data.loc[mask & (data['Canceled_Rides_by_Customer'] == 'AC is Not working '), 'Incomplete_Rides_Reason'] = 'AC is Not working '
data.loc[mask & (data['Canceled_Rides_by_Customer'] == 'Wrong Address'), 'Incomplete_Rides_Reason'] = 'Wrong Address by customer'

# Now using feature 'Canceled_Rides_by_Driver':
data.loc[mask & (data['Canceled_Rides_by_Driver'] == 'Personal & Car related issue '), 'Incomplete_Rides_Reason'] = 'Personal & Car related issue of Driver'
data.loc[mask & (data['Canceled_Rides_by_Driver'] == 'Customer related issue '), 'Incomplete_Rides_Reason'] = 'Driver had Customer related issue '
data.loc[mask & (data['Canceled_Rides_by_Driver'] == 'Customer was coughing/sick'), 'Incomplete_Rides_Reason'] = 'Customer was coughing/sick'
data.loc[mask & (data['Canceled_Rides_by_Driver'] == 'More than permitted people in there '), 'Incomplete_Rides_Reason'] = 'More than permitted people in there '



```
